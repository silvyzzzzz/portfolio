## Merapikan Dataframe

```python
import numpy as np
#Slicing isi dataframe yang diperlukan
df = data.iloc[2:,0:]
df = df.drop(np.array(range(36,41)))
#Mengganti nama kolom data frame
df = df.reset_index()
df.drop(columns=["index"], inplace=True)
df = df.rename(columns=dict(zip(df.columns[1:], np.array(data.iloc[0, 1:]))))

```

```python
#Mengecek dimensi data frame
df.shape
```

```python
#Mengganti isi "-" menjadi nan
df.replace("-", np.nan, inplace=True)
```

```python
#Mengecek missing value
df.isna().sum()
```

Dari output di atas, terdapat beberapa kolom yang tidak memiliki isi sama sekali (berisi missing value semua) sehingga variabel-variabel tersebut dapat dihapus

```python
df.drop(columns=["Anyelir (Tangkai)", "Gladiol (Tangkai)", "Melati (Pohon)"], inplace=True)
```

```python
df.info()
```

```python
df
```

```python
#Mengubah jenis value yang semula berjenis object atau string menjadi float
for i in df.columns[1:]:
    df[i]=df[i].astype(float)
```

```python
#Mengisi missing value dengan mean
for i in df.columns[1:]:
    df[i] = df[i].fillna(df[i].mean())
```

```python
#Mengecek apakah masih ada missing value
df.isna().sum()
```

```python
#EDA
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 30))
for i in range(0, len(df.columns[1:])):
    plt.subplot(8, 2,i+1)
    sns.distplot(df[df.columns[1:][i]], color='orange')
    plt.tight_layout()
```

Dari grafik distribusi di atas, dapat terlihat bahwa semuanya tidak berdistribusi normal sehingga perlu dilakukan normalisasi

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df1 = df.copy()
df1[df1.columns[1:]] = scaler.fit_transform(df1[df1.columns[1:]])
```

# Clustering on Province

```python
from sklearn.cluster import KMeans
inertia = []

for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df1.iloc[0:,1:])
    inertia.append(kmeans.inertia_)



plt.figure(figsize=(20, 10))

sns.lineplot(x=range(2, 11), y=inertia, color='#000087', linewidth = 4)
sns.scatterplot(x=range(2, 11), y=inertia, s=300, color='#800000',  linestyle='--')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(df1.iloc[0:,1:])
df['cluster'] = kmeans.labels_
```

```python
#CLuster 0
df[df['cluster']==0]
```

```python
#Cluster 1
df[df['cluster']==1]
```

```python
df[df['cluster']==2]
```

```python
summary_province = df[df.columns[1:].tolist()].groupby('cluster').mean()
```

```python
summary_province
```

```python
# Mencari rata-rata pada setiap baris
rata_rata = np.mean(summary_province, axis=1)

# Mengurutkan rata-rata dari yang tertinggi
indeks_terurut = np.argsort(rata_rata)[::-1]
rata_rata_terurut = rata_rata[indeks_terurut]

rata_rata_terurut
```

## Clustering on flower

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2 = data_clean.transpose()
df2.set_axis(df2.iloc[0], axis='columns', inplace=True)
df2 = df2[1:]  # Remove the first row
df2.reset_index(inplace=True)
df2.rename(columns={'index': 'Bunga'}, inplace=True)
df2_clean = df2.copy()
df2[df2.columns[1:]] = scaler.fit_transform(df2[df2.columns[1:]])
```

```python
from sklearn.cluster import KMeans
inertia = []

for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df2.iloc[0:,1:])
    inertia.append(kmeans.inertia_)



plt.figure(figsize=(20, 10))

sns.lineplot(x=range(2, 11), y=inertia, color='#000087', linewidth = 4)
sns.scatterplot(x=range(2, 11), y=inertia, s=300, color='#800000',  linestyle='--')
plt.show()
```

```python
kmeans = KMeans(n_clusters=7, random_state=0).fit(df2.iloc[0:,1:])
df2_clean['cluster'] = kmeans.labels_
```

```python
for i in range(0,7):
    print("Cluster", i, "\n", df2_clean[df2_clean['cluster']==i]["Bunga"], "\n")
```

```python
summary_flower= df2_clean[df2_clean.columns[1:].tolist()].groupby('cluster').mean()
```

```python
summary_flower
```

```python
# Mencari rata-rata pada setiap baris
rata_rata = np.mean(summary_flower, axis=1)

# Mengurutkan rata-rata dari yang tertinggi
indeks_terurut = np.argsort(rata_rata)[::-1]
rata_rata_terurut = rata_rata[indeks_terurut]

rata_rata_terurut
```
